```python
df['delivery_SMA_5'] = df.iloc[:,13].rolling(window=5).mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=[15,10])
df.dropna(inplace=True)
plt.grid(True)
df.index = df.Date
#plt.xticks(df['Date'])
plt.plot(df['Deliverable Qty'],label='data')
plt.plot(df['delivery_SMA_3'],label='SMA 3 Months')
plt.plot(df['delivery_SMA_5'],label='SMA 5 Months')
plt.legend(loc=2)
```
